𝙑𝙄𝙐𝙎𝘼𝙇𝙄𝙕𝙄𝙉𝙂 𝙏𝙃𝙀 𝘿𝘼𝙏𝘼

𝙁𝘼𝙈𝙄𝙇𝙔 𝙑𝙎 𝙁𝙀𝙈𝘼𝙇𝙀 𝙈𝘼𝙏𝙐𝙍𝙄𝙏𝙔 𝘿𝘼𝙔𝙎 𝘼𝙉𝘿 𝙁𝘼𝙈𝙄𝙇𝙔 𝙑𝙎 𝙈𝘼𝙇𝙀 𝙈𝘼𝙏𝙐𝙍𝙄𝙏𝙔 𝘿𝘼𝙔𝙎

```python
a=data[["Family","Female maturity (days)"]].groupby(["Family"],axis=0).sum().sort_values(by="Female maturity (days)",ascending=False).head(10)
a
```

```python
b=data[["Family","Male maturity (days)"]].groupby(["Family"],axis=0).sum().sort_values(by="Male maturity (days)",ascending=False).head(10)
b
```

```python
fig=plt.figure()
ax0=fig.add_subplot(1,2,1)
ax1=fig.add_subplot(1,2,2)


a.plot(kind="line",color="Orange",figsize=(15,5), ax=ax0)
ax0.set_title("TOP 10 FAMILY WITH THE MOST FEMALE MATURITY RATE",fontsize=13)
ax0.set_xlabel("FAMILY",fontsize=12)
ax0.set_ylabel("FEMALE MATURITY DAYS",fontsize=12)


b.plot(kind="line",color="Blue",figsize=(15,5), ax=ax1)
ax1.set_title("TOP 10 FAMILY WITH THE MOST MALE MATURITY RATE", fontsize=13)
ax1.set_xlabel("FAMILY",fontsize=12)
ax1.set_ylabel("MALE MATURITY DAYS",fontsize=12)

plt.show()
```

```python
c.plot(kind="bar",color="Green",figsize=(13,5),fontsize=9)
plt.title("TOP 10 FAMILY INCUBATION DAYS",fontsize=14)
plt.xlabel("FAMILY",fontsize=12)
plt.xticks(fontsize=12)
plt.ylabel("INCUBATION DAYS",fontsize=12)
plt.show()
```
